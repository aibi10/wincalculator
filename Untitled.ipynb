```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians', 'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',
       'Delhi Capitals']
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team1'].isin(teams)]
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id', 'winner', 'total_runs']]
```

```python
match_df
```

```python
delivery['batting_team'].unique()
```

```python
delivery['batting_team'] = delivery['batting_team'].replace('Deccan Chargers', 'Sunrisers Hyderabad')
delivery['batting_team'] = delivery['batting_team'].replace('Delhi Daredevils', 'Delhi Capitals')

delivery['bowling_team'] = delivery['bowling_team'].replace('Deccan Chargers', 'Sunrisers Hyderabad')
delivery['bowling_team'] = delivery['bowling_team'].replace('Delhi Daredevils', 'Delhi Capitals')
```

```python
delivery['batting_team'].unique()
```

```python
delivery = delivery[delivery['batting_team'].isin(teams)]
delivery = delivery[delivery['bowling_team'].isin(teams)]
```

```python
match_df
```

```python
delivery_df = match_df.merge(delivery, on = 'match_id')
```

```python
delivery_df
```

```python
delivery_df.shape
```

```python
delivery_df['inning'].value_counts()
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.inning.value_counts()
```

```python
delivery_df
```

```python
delivery_df['current_score'] = delivery_df.groupby(['match_id']).cumsum()['total_runs_y']
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over'] - 1) * 6 - delivery_df['ball']
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna(0)
```

```python
def conversion(player_dismissed):
    if player_dismissed != 0:
        player_dismissed = 1
    return player_dismissed
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(conversion)
```

```python
delivery_df['player_dismissed'].value_counts()
```

```python
delivery_df
```

```python
delivery_df['wickets1'] = delivery_df.groupby(['match_id']).cumsum()['player_dismissed']
```

```python
delivery_df['wickets'] = 10 - delivery_df['wickets1']
```

```python
delivery_df
```

```python
delivery_df['crr'] = delivery_df['current_score'] * 6/ (120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = delivery_df['runs_left'] * 6/ delivery_df['balls_left']
```

```python
delivery_df = delivery_df[delivery_df['balls_left'] != 0]
```

```python
def res(result):
    if result['batting_team'] == result['winner']:
        return 1
    else:
        return 0
```

```python
delivery_df['result'] = delivery_df.apply(res, axis = 1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
X = final_df.iloc[ : ,  : -1]
y = final_df.iloc[ : , -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse = False, drop = 'first'),['batting_team', 'bowling_team'])
]
, remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
